```python
import re as regex
```

```python

with open('data.txt','r') as f:
  contents = f.read()
  pattern = regex.compile(r'\d\d\d.\d\d\d.\d\d\d')

  matches = pattern.finditer(contents)

  for match in matches:
    print(match)

  # print(contents)
```

```python
with open('data.txt','r') as f:
  contents = f.read()
  pattern = regex.compile(r'\d\d\d[-.]\d\d\d[-.]\d\d\d')

  matches = pattern.finditer(contents)

  for match in matches:
    print(match)
```

```python
with open('data.txt','r') as f:
  contents = f.read()
  pattern = regex.compile(r'[89]00[-.]\d\d\d[-.]\d\d\d')

  matches = pattern.finditer(contents)

  for match in matches:
    print(match)
```

```python
with open('data.txt','r') as f:
  contents = f.read()
  pattern = regex.compile(r'\d{3}[-.]\d{3}[-.]\d{4}')

  matches = pattern.finditer(contents)

  for match in matches:
    print(match)
```

```python
pattern = regex.compile(r'Mr\.?\s[A-Z]')

matches = pattern.finditer(text_to_search)

for match in matches:
  print(match)
```

```python
pattern = regex.compile(r'Mr\.?\s[A-Z]\w*')

matches = pattern.finditer(text_to_search)

for match in matches:
  print(match)
```
